# Import Data

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

# District Summary

```python
#create columns
totSchools = df.school_name.nunique()
totStudents = df["Student ID"].nunique()
totBudget = df.groupby("school_name").budget.mean().sum()
avgMath = df.math_score.mean()
avgReading = df.reading_score.mean()

mask1 = df.math_score >= 70
percMath = 100 * df.loc[mask1, ["Student ID"]].count().values[0] / df["Student ID"].nunique()

mask2 = df.reading_score >= 70
percReading = 100 * df.loc[mask2, ["Student ID"]].count().values[0] / df["Student ID"].nunique()

mask3 = (df.math_score >= 70) & (df.reading_score >= 70)
percBoth = 100 * df.loc[mask3, ["Student ID"]].count().values[0] / df["Student ID"].nunique()

#create table
districtSummary = pd.DataFrame()
districtSummary["Total Schools"] = [totSchools]
districtSummary["Total Students"] = [totStudents]
districtSummary["Total Budget"] = [totBudget]
districtSummary["Average Math Score"] = [avgMath]
districtSummary["Average Reading Score"] = [avgReading]
districtSummary["% Passing Math"] = [percMath]
districtSummary["% Reading Score"] = [percReading]
districtSummary["% Overall Passing"] = [percBoth]

#print
districtSummary
```
